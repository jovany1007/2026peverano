# RAICES


```python
def tipo_raices(a, b, c):
    discriminante = b**2 - 4*a*c
    
    print(f"Discriminante = {discriminante}")
    
    if discriminante < 0:
        print("Sus raíces son complejas")
    elif discriminante == 0:
        print("Sus raíces son reales e iguales")
    else:
        print("Sus raíces son reales y diferentes")
```

```python
tipo_raices(1/3,-2,4)
```
